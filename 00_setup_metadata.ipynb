```python
# Setup Parameters 
CATALOG = "dbw_chinook_team"
BRONZE  = f"{CATALOG}.chinook_bronze"

print(f"Catalog : {CATALOG}")
print(f"Bronze  : {BRONZE}")
```

```python
# ── Create Parent Metadata Table ───
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {BRONZE}.pipeline_metadata (
    table_name    STRING,
    file_name     STRING,
    active_flag   STRING,
    created_date  TIMESTAMP,
    modified_date TIMESTAMP
)
""")

spark.sql(f"TRUNCATE TABLE {BRONZE}.pipeline_metadata")

spark.sql(f"""
INSERT INTO {BRONZE}.pipeline_metadata VALUES
('Artist',       'artist.parquet',        'Y', current_timestamp(), current_timestamp()),
('Album',        'album.parquet',         'Y', current_timestamp(), current_timestamp()),
('Track',        'track.parquet',         'Y', current_timestamp(), current_timestamp()),
('Genre',        'genre.parquet',         'Y', current_timestamp(), current_timestamp()),
('MediaType',    'mediatype.parquet',     'Y', current_timestamp(), current_timestamp()),
('Customer',     'customer.parquet',      'Y', current_timestamp(), current_timestamp()),
('Employee',     'employee.parquet',      'Y', current_timestamp(), current_timestamp()),
('Invoice',      'invoice.parquet',       'Y', current_timestamp(), current_timestamp()),
('InvoiceLine',  'invoiceline.parquet',   'Y', current_timestamp(), current_timestamp()),
('Playlist',     'playlist.parquet',      'Y', current_timestamp(), current_timestamp()),
('PlaylistTrack','playlisttrack.parquet', 'Y', current_timestamp(), current_timestamp())
""")

print(f"✅ Parent metadata table created")
spark.table(f"{BRONZE}.pipeline_metadata").show()
```

```python
#  Create Child Execution Log Table 
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {BRONZE}.pipeline_execution_log (
    table_name       STRING,
    execution_time   TIMESTAMP,
    status           STRING,
    source_row_count LONG,
    target_row_count LONG,
    file_location    STRING,
    created_date     TIMESTAMP
)
""")

print("✅ Execution log table created")
print("\n=== SETUP COMPLETE ===")
print(f"  {BRONZE}.pipeline_metadata")
print(f"  {BRONZE}.pipeline_execution_log")
```
